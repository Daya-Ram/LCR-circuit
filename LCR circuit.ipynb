##### Resonant frequency 

$f_0 = \frac{1}{2\pi\sqrt(LC)}$

#### Quality factor

Q = 1/R $\sqrt{L/C})$

#### Inductive reactance

$X_L = 2\pi fL$

#### Capacitive reactance

$X_c = \frac{1}{2\pi fC}$

#### Impedance

Z = $\sqrt{(X_L - X_C)^2 + R^2}$

#### Phase angle

$\phi = tan^-1\frac{(X_L - X_C)}{R})$

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
class lcr():
    def __init__(self,L,C,R,f):
        self.L = L
        self.C = C
        self.R = R
        self.f = f
        
    def f0(self):
        fr = (2*np.pi*np.sqrt(self.L*self.C))**-1
        return fr
    
    def Q(self):
        qf = np.sqrt(self.L/self.C)/self.R
        return qf
    
    def inductiveReactance(self):
        XL = 2*np.pi*self.L*self.f
        return XL
    
    def capacitiveReactance(self):
        XC=1/(2*np.pi*self.C*self.f)
        return XC 
    
    def impedance(self):
        im=np.sqrt((self.inductiveReactance()-self.capacitiveReactance())**2+self.R**2)
        return im
    
    def phase(self):
        pi=np.arctan((self.inductiveReactance()-self.capacitiveReactance())/self.R)*180/np.pi
        return pi    # degree
        
```

```python
l = 0.5       
c = 10**-6
r = 50
v = 220
```

```python
a = lcr(l,c,r,220)
print('Resonating frequency:',a.f0(),'Hz')
print('Q factor:',a.Q())
```

```python
freq=np.arange(20,500,10)
b=lcr(l,c,r,freq)
xl=b.inductiveReactance()
xc=b.capacitiveReactance()
z=b.impedance()
ph=b.phase()
I=v/z  # current
vr=I*r  #p.d.across reisitor
vl=I*xl  #p.d.across inductor
vc=I*xc  #p.d.across capacitor
power =v*I*np.cos(ph*np.pi/180)
```

```python
plt.plot(freq,xl,'b_',label='Inductive Reactance')
plt.plot(freq,xc,'r.',label='Capacitive Reactance')
plt.xlabel(xlabel="Frequency(Hz)",fontsize=14,fontstyle='italic',color="black")
plt.ylabel(ylabel="Reactance(ohm)",fontsize=14,fontstyle='italic',color="black")
plt.title(label="frequency vs Reactance curve",fontsize=20,fontweight='bold',color="red")
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.savefig("LCR1.pdf", dpi = 500) # dpi dot per inch
plt.show()
plt.close()
```
